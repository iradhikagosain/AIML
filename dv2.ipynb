```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
```

```python
tips
```

```python
sns.set_theme(style="ticks")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
flights=sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
    
)
```
