```python
class BankAccount:
    def __init__(self,oname,anum,bal):
        self.oname=oname
        self.anum=anum
        self.bal=bal
    def deposite_money(self,depo):
        self.bal=self.bal+depo
        print(self.bal)
    def withdraw_money(self,draw):
        self.bal=self.bal-draw
        print(f"withdrawl money is {draw} and remaning balance is {self.bal}")

b1=BankAccount("radhika",12344839,45_000)
print(b1)
print(b1.oname)
print(b1.anum)
print(b1.bal)
b1.deposite_money(40000)
print(b1.bal)
b1.withdraw_money(5000)
print(b1.bal)
```

```python
class Book:
    
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.list_of_reviews = []   

    def add_review(self, review):
        self.list_of_reviews.append(review)

    def count_reviews(self):
        print("No. of reviews:", len(self.list_of_reviews))

    def display_reviews(self):
        print("Reviews:")
        for review in self.list_of_reviews:
            print("-", review)



c1 = Book("Atomic Habits", "James Clear")

print(c1.title)
print(c1.author)

c1.add_review("Great book")
c1.add_review("Very motivating")
c1.add_review("Must read")

c1.count_reviews()
c1.display_reviews()
```

```python
class Student:

    def __init__(self, name, roll_no, marks):
        self.set_name(name)
        self.set_roll_no(roll_no)
        self.set_marks(marks)

    def set_marks(self, marks):
        if marks < 0:
            print("Marks cannot be negative")
        else:
            self._marks = marks
            
    def get_marks(self):
        return self._marks

    def set_roll_no(self, roll_no):
        if 1 <= roll_no <= 100:
            self._roll_no = roll_no
        else:
            print("Roll number must be between 1 and 100")

    def get_roll_no(self):
        return self._roll_no

    def set_name(self, name):
        if name.strip() == "":
            print("Name cannot be empty")
        else:
            self._name = name
            
    def get_name(self):
        return self._name


s1 = Student("Riya", 67, 89)

print("Name:", s1.get_name())
print("Roll No:", s1.get_roll_no())
print("Marks:", s1.get_marks())
```

```python
import math
class shape:
    def area(self):
        print("area of the shape")


class circle(shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        print("area of circle:",math.pi*self.r*self.r)

class rectangle(shape):
        def __init__(self,l,b):
            self.l=l
            self.b=b
        def area(self):
            print("area of rectangle:",self.l*self.b)

class triangle(shape):
    def __init__(self,bs,h):
        self.bs=bs
        self.h=h
    def area(self):
        print("area of triangle:",0.5*self.bs*self.h)


c=circle(6)
r=rectangle(46,74)
t=triangle(23,4)

c.area()
r.area()
t.area()

    
        
```

```python
class Employee:
    def __init__(self, salary):
        self.salary = salary

    def calculate_salary(self):
        pass


class Intern(Employee):
    def __init__(self, salary, stipend):
        super().__init__(salary)
        self.stipend = stipend

    def calculate_salary(self):
        total = self.salary + self.stipend
        print("Intern Salary =", total)


class FullTimeEmployee(Employee):
    def calculate_salary(self):
        print("Full Time Salary =", self.salary)


class ContractEmployee(Employee):
    def __init__(self, salary, months):
        super().__init__(salary)
        self.months = months

    def calculate_salary(self):
        total = self.salary * self.months
        print("Contract Salary =", total)


i = Intern(10000, 5500)
f = FullTimeEmployee(40000)
c = ContractEmployee(55000, 3)


i.calculate_salary()
f.calculate_salary()
c.calculate_salary()
```

```python
class Herbivore:
    def __init__(self, plant, habit):
        self.plant = plant
        self.habit = habit

    def behaviour(self):
        print("Eats only plants")


class Carnivore:
    def __init__(self, meat, hunting):
        self.meat = meat
        self.hunting = hunting

    def behaviour(self):
        print("Eats only animals")


class Omnivore:
    def __init__(self, pfood, afood):
        self.pfood = pfood
        self.afood = afood

    def behaviour(self):
        print("Eats both plants and animals")


class Bear(Herbivore, Carnivore, Omnivore):
    def __init__(self, plant, habit, meat, hunting, pfood, afood, name):
        Herbivore.__init__(self, plant, habit)
        Carnivore.__init__(self, meat, hunting)
        Omnivore.__init__(self, pfood, afood)
        self.name = name



b = Bear("berries", "slow eater", "deer meat", "hunts", "fruits", "fish", "Cutu")

print(b.name)
print(b.plant, b.habit)
print(b.meat, b.hunting)
print(b.pfood, b.afood)
```
