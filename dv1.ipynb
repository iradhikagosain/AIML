```python
oscar_movies=[
    "the dark knight",
    "the hurt locker",
    "the king's speech",
    "the artist",
    "argo"
    
]

years=[2008,2009,2010,2011,2012]
oscar_revenue=[1005,170,427,133,232]

plt.plot(years,oscar_revenue)
plt.title("oscar movie revenue")
plt.xlabel("years")
plt.ylabel("oscar revenue")
```

```python
non_oscar_movies=["slumdog","avatar","inception","hugo","lincoln"]
non_oscar_years=[2008,2009,2010,2011,2012]
non_oscar_revenue=[378,2788,829,185,275]
```

```python
plt.plot(years,non_oscar_revenue,label="non oscar")
plt.plot(years,oscar_revenue,label="oscar")

plt.title("oscar vs non oscar")
plt.xlabel("year")
plt.ylabel("revenue")
plt.legend(loc="lower left")
```

```python
plt.plot(years,non_oscar_revenue,"d:b",label="non oscar")
plt.plot(years,oscar_revenue,color="red",marker="o",linewidth=3,label="oscar")

plt.title("oscar vs non oscar")
plt.xlabel("year")
plt.ylabel("revenue")
plt.legend(loc="lower left")
```

```python
# comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

# Plotting Bar Chart
import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar")
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
plt.tight_layout()

plt.xticks(x, years) # to show x-axis labels
```
