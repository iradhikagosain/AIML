```python
def calculator(a,b,operation):
    if operation=='+':
        print(a+b)
    elif operation=='-':
        print(a-b)
    elif operation=='*':
        print(a*b)
    elif operation=='/':
        print(a/b)
    else:
        print("NOT APPLICABLE")
```

```python
calculator(3,76,"+")
```

```python
def is_prime(n):
    if(n<2):
        print("not a prime number")
    else:
        for i in range(2,n-1):
            if  n%i==0:
                print("not a prime numeber")
    print("prime number")

is_prime(9)
```
