```python
from sklearn.svm import SVR
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,classification_report
```

```python
df=datasets.load_diabetes(as_frame=True).frame
```

```python
X=df.drop("target",axis=1)
Y=df["target"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(
    X,Y,test_size=0.3,random_state=42
)
```

```python
Y_scaler = StandardScaler()
Y_train_scaled=Y_scaler.fit_transform(Y_train.values.reshape(-1,1)).ravel()
Y_test_scaled=Y_scaler.transform(Y_test.values.reshape(-1,1)).ravel()
```

```python
svr=SVR()
svr.fit(X_train,Y_train_scaled)
```

```python
Y_pred_scaled=svr.predict(X_test)
```

```python
print("R2:",r2_score(Y_test_scaled,Y_pred_scaled))
```

```python
#hyper parameter tuning
from sklearn.model_selection import GridSearchCV

```

```python
param_grid={
    "C":[1,2,4,6,8,21],
    "kernel":["rbf","linear","poly","sigmoid"],
    "epsilon":[0.01,0.1,0.2,0.3,0.4]
}
```

```python
svr=SVR()
grid_search=GridSearchCV(svr,param_grid,scoring="r2",cv=5)
grid_search.fit(X_train,Y_train_scaled)
```

```python
print("best params:",grid_search.best_params_)
```

```python
best_model=SVR(kernel="linear",C=21,epsilon=0.1)
best_model.fit(X_train,Y_train_scaled)

Y_pred_scaled=best_model.predict(X_test)
print("R2:",r2_score(Y_test_scaled,Y_pred_scaled))
```

```python
from sklearn.svm import LinearSVR
svr=LinearSVR(C=10,epsilon=0.1,max_iter=10000)
svr.fit(X_train,Y_train_scaled)

Y_pred_scaled=svr.predict(X_test)
print("R2:",r2_score(Y_test_scaled,Y_pred_scaled))
```
