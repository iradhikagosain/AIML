```python
import json
data={
    "mumbai":89330,
    "delhi":2383,
    "goa":3334
}
with open("data.json","w") as f:
    json.dump(data,f,indent=4)

new_city=input("enter city name:")
population=int(input("enter population:"))
data[new_city]=population

with open("data.json","w")as f:
    json.dump(data,f)

for city,population in data.items():
    print(city,":",population)
```
