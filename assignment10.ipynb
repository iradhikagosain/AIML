```python
import requests
from bs4 import BeautifulSoup
import time

page = 1

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

while True:
    URL = f"https://quotes.toscrape.com/page/{page}/"

    try:
        response = requests.get(URL, headers=headers, timeout=10)
    except requests.exceptions.RequestException as e:
        print(f"Request failed on page {page}: {e}")
        break

    soup = BeautifulSoup(response.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No more pages. Scraping finished.")
        break

    with open(f"scraped_data/quotes{page}.html", "w", encoding="utf-8") as f:
        f.write(response.text)

    print(f"Downloaded page {page}")

    page += 1
    time.sleep(1)   

    
```

```python

for i in range(1,10):
    with open(f"scraped_data/quotes{i}.html","r",encoding="utf-8")as f:
        content=f.read()
        print(f"data of page{i}")
        


    soup=BeautifulSoup(content,"lxml")
    all_quotes = soup.select("div.quote")
    life_quotes = []

    for q in all_quotes:
        all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

    print(life_quotes)
   
    
```
