```python
import json
with open("store_data.json","r")as f:
    data=json.load(f)

print(data)
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    
    for user in data:
        raw_rating=str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]
        user["rating"]=raw_rating

        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None

        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)


```

```python
def get_insights(data):
    total_rating=0
    for user in data:
        total_rating+=float(user["rating"])

    print("average is :",total_rating/len(data))

    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1
            
    print("percentage is :",(poor_rating/len(data))*100)       
```

```python
get_insights(data)
```

```python
def get_recommendation(data):
    recommendations=[]

    for user in data:
        curr_recom={}
        curr_recom["name"]=user["name"]
        if float(user["rating"])>=4:
            curr_recom["brand"]="apple"
        else:
            curr_recom["brand"]="samsung"

        recommendations.append(curr_recom)
    return recommendations
            
```

```python
get_recommendation(data)
```
