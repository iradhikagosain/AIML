```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, classification_report, confusion_matrix
```

```python
df=pd.read_csv("shop_smart_ecommerce.csv")
```

```python
X=df.drop(columns=["Revenue"],axis=1)
Y=df["Revenue"]
```

```python
df.info()
```

```python
num_features=X.select_dtypes(include=["int64","float64"])
col_features=X.select_dtypes(include=["object","bool"])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)
```

```python
ct = ColumnTransformer(
   transformers=[
        ("num", StandardScaler(), num_features),
        ("col", OneHotEncoder(handle_unknown="ignore"), col_features)
    ])
```

```python
dt = DecisionTreeClassifier(
    max_depth=6,              # prevents deep overfitting
    min_samples_leaf=30,      # smooths decision boundaries
    class_weight="balanced",  # handles imbalance
    random_state=42
)
```

```python
pipe = Pipeline(
    steps=[
        ("preprocess", ct),
        ("model", dt)
    ]
)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred = pipe.predict(X_test)

print("F1 score:", f1_score(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
```
