```python
import pandas as pd
```

```python
#data structure
#series(1d)and dataframe(2d)

info={
"name":["riya","mahi","neha"],
"cgpa":[4.6,9.0,7.6]
}
df=pd.DataFrame(info)
print(df,type(df))


df1=pd.Series(info)
print(df1,type(df1))
print(df1[0])
print(df1[1])

```

```python
#labels

s=pd.Series([34,2,4,55,22],index={"riya","babbu","mahi","diksha","pooja"})
print(s)

print(s["diksha"])
print(s.index)
```

```python
#vectorization
s1=pd.Series([1,4,6,7,7])
s2=pd.Series([1,4,6,7,7])
s1[0]=100
print(s1)
s1.add(88)
print(s1)
print(s1+s2)
```
